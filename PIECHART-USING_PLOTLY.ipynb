```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv('bird.csv')
data
```

```python
figure = px.bar(data, x='Bldg #', y='Deaths', title='Birds Death Against Buildings', color_discrete_sequence=['maroon'])
figure.show()
```

```python
figure2 = px.pie(data, names='Common Name', hole=0.4, title='Donut Pie Chart')

colors = px.colors.qualitative.Set3
figure2.update_traces(marker=dict(colors=colors))
figure2.update_layout(
    width=800,
    height=800,
    showlegend=True,
    legend=dict(title='Bird Species'),
    font=dict(family='Arial', size=12),
)

figure2.show()
```

```python
figure3 = px.pie(data, names='Common Name', color='Side', title='Pie Chart with Building Sides Color')

figure3.update_layout(width=1000, height=1000)

figure3.show()

```

```python
fig_pie_go = go.Figure(go.Pie(labels=data["Common Name"], values=data["Deaths"]))

fig_pie_go.update_traces(textfont=dict(color='maroon'))

fig_pie_go.update_layout(width=1000, height=1000)

fig_pie_go.show()
```

```python
fig_pie_hover = go.Figure(go.Pie(labels=data["Bldg #"], values=data["Deaths"], hoverinfo='label+percent'))


fig_pie_hover.update_layout(width=1000, height=1000)

fig_pie_hover.show()
```

```python
fig_no_click = go.Figure(go.Pie(labels=data["Common Name"], values=data["Deaths"]))

fig_no_click.update_layout(height = 1000, width = 1000)

fig_no_click.show(config={'scrollZoom': False, 'displayModeBar': False})
```
